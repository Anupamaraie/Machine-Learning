# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
```

# Read dataset

Since, there can only be 36 runs at max in 1 over, the score exceeding 36 is anomaly. 

```python
df = pd.read_csv('cricket.csv')
df
```

# Model Training

```python
model = IsolationForest(n_estimators=100, max_samples = 'auto', contamination = float(0.2), max_features=1.0)
model.fit(df[['Scores']])
```

# Predictions

```python
df['anomalies_scores'] = model.decision_function(df[['Scores']])
x=df['anomalies_scores']
df['anomaly'] =model.predict(df[['Scores']])
y=df['anomaly']
df.head(20)
```

```python
plt.scatter(y,df['Overs'], marker='x')
plt.xlabel('Anomaly')
plt.ylabel('Overs')

plt.axis([-2,2,0,25])
```
